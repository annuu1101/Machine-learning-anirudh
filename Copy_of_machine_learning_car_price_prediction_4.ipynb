```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('/content/Car_data')
df.head()
```

```python
input=df.drop(['Unnamed: 0','Price'],axis=1)
input
```

```python
target=df['Price']
target
```

```python
from sklearn.model_selection import train_test_split
train_test_split
```

```python
# prompt: train_test_split

X_train, X_test, y_train, y_test = train_test_split(input, target, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
ohe=OneHotEncoder()
ohe.fit_transform(input[['name','company','fuel_type']])
```

```python
ohe.categories_
```

```python
ct = make_column_transformer(
    (OneHotEncoder(categories=ohe.categories_), ['name','company','fuel_type']),
    remainder='passthrough')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.pipeline import make_pipeline
pipline_log=make_pipeline(ct,LinearRegression())
```

```python
pipline_log
```

```python
pipline_log.fit(X_train,y_train)
```

```python
y_pred=pipline_log.predict(X_test)
y_pred
```

```python
pipline_log.score(X_test,y_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred).round(2)
```
